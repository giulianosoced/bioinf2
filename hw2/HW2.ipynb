```python
#we had downloaded the Curated gene list for disease C1621958 from disgenet. let's open the file
disgenet=pd.read_csv('browser_source_summary_gda_CURATED.tsv', sep="\t")
```

```python
##let's see how it looks like
# disgenet
```

```python
#now let's save the gene symbols, entrez ID and names in arrays
symbol=[]
names=[]
entrezID=[]
for i in range(len(disgenet)):
#     print(disgenet['Gene'][i]+" ")
    symbol.append(disgenet['Gene'][i])
    names.append(disgenet['Gene_Full_Name'][i])
    entrezID.append(disgenet['Gene_id'][i])
    
#we check on HGNC to see if we need to change genes name 
```

```python
# we need to change
# H2AFX H2AX
# H3F3A H3-3A
# HIST1H3A H3C1 
# HIST1H3B H3C2 
# SEPT14 SEPTIN14

for i in range(len(symbol)):
    if symbol[i]=='H2AFX':
        symbol[i]='H2AX'
    if symbol[i]=='H3F3A':
        symbol[i]='H3-3A'
    if symbol[i]=='HIST1H3A':
        symbol[i]='H3C1'
    if symbol[i]=='HIST1H3B':
        symbol[i]='H3C2'
    if symbol[i]=='SEPT14':
        symbol[i]='SEPTIN14'
#     print(symbol[i]+ ",")
```

```python
#with our genes name fixed we have seen the list of uniprot genes ID and saved it
uniprot=pd.read_csv('download.csv',sep=";")
```

```python
## it looks like this
# uniprot
```

```python
#let's save that IDs in array.
uniprotAC=[]
for i in range(len(uniprot)):
#     print(uniprot['Entry'][i]+" ")
    uniprotAC.append(uniprot['Entry'][i])
```

```python
#ADD THE DESCRIPTION COLUMN!
```

```python
#save our data to a new df
results = pd.DataFrame(list(zip(symbol,names,entrezID,uniprotAC)), columns=['Symbol','Name','ID','UniprotAC'])
results.to_csv('Results.csv')
```

```python
#open biogrid DB
biogrid=pd.read_csv('BIOGRID-ALL-3.5.179.txt', sep='\t')
```

```python
 biogrid
```

```python
#let's select what we want from the biogrid dataset.
#select only human PPI
biogrid_human=biogrid.loc[(biogrid['ORGANISM_A_ID']==9606) & (biogrid['ORGANISM_B_ID']==9606)]
biogrid_human
```

```python
#select only interaction on our seed genes
biogrid_seed=biogrid_human.loc[(biogrid_human['OFFICIAL_SYMBOL_A'].isin(symbol)) | (biogrid_human['OFFICIAL_SYMBOL_B'].isin(symbol))]
biogrid_seed
```

```python
#now let's search fot genes not in our list, they may have interaction among them.
#save them into a list

#select all interaction where one of two interactor is not one seed gene
non_seed_df=biogrid_seed.loc[~(biogrid_seed['OFFICIAL_SYMBOL_A'].isin(symbol)) | (biogrid_seed['OFFICIAL_SYMBOL_B'].isin(symbol))]
#reset index
non_seed_df=non_seed_df.reset_index(drop=True)
#create empty list
non_seed_list=[]
#fill the list
for i in range(len(non_seed_df)):
    if non_seed_df['OFFICIAL_SYMBOL_A'][i] not in symbol:
        non_seed_list.append(non_seed_df['OFFICIAL_SYMBOL_A'][i])
    if non_seed_df['OFFICIAL_SYMBOL_B'][i] not in symbol:
        non_seed_list.append(non_seed_df['OFFICIAL_SYMBOL_B'][i])
```

```python
#now search for non seed interactions from the human DB

#reset index
biogrid_human=biogrid_human.reset_index(drop=True)
#create a list of index of the original matrix with non seed interaction to save
idx=[]
#fill the list
for i in range(len(biogrid_human)):
    if biogrid_human['OFFICIAL_SYMBOL_A'][i] in non_seed_list and biogrid_human['OFFICIAL_SYMBOL_B'][i] in non_seed_list:
        idx.append(i)
```
